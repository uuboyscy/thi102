```python
import glob

import pandas as pd
```

```python
data = []

for path in glob.glob("res_gossiping/*.txt"):
    with open(path, "r", encoding="utf-8") as f:
        article_content_list = f.read().split("---split---")
        if len(article_content_list) != 2:
            print(path)
            continue

        row = article_content_list[1].split("\n")[1:-1]
        # print(row)
        data.append(row)

data
```

```python
data[0]
```

```python
columns = [col_str.split(": ")[0] for col_str in data[0]]
columns
```

```python
df = pd.DataFrame(
    data=data,
    columns=columns,
)

df
```

```python
df["推"].apply(
    lambda x: int(x.split(": ")[1])
)
```

```python
for col in df:
    print(col)
```

```python
for col in df:
    df[col] = df[col].apply(
        lambda x: x.split(": ")[1]
    )
```

```python
df["推"] = df["推"].astype(int)
```
